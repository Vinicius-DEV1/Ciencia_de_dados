# PHQ9

```python
dataset3
```

# PHQ_GRP

Por fim, crie a variável phq_grp - grupo de sintomas de depressão que assume 0 (“sem sintomas”) se phq9 < 5, 1 (“sintomas leves”) se 5 <= phq9 < 10, 2 (“sintomas moderados”) se 10 <= phq9 < 15, 3 (“sintomas moderadamente severos”) se 15 <= phq9 < 19 e 4 (“sintomas severos”) se phq9 >= 20

```python
dataset3['phq9'].unique()
```

```python
dataset3.loc[dataset3['phq9']<5,'phq_grp'] = 0
dataset3.loc[(dataset3['phq9']>=5) & (dataset3['phq9']<10),  'phq_grp'] = 1
dataset3.loc[(dataset3['phq9']>=10) & (dataset3['phq9']<15), 'phq_grp'] = 2
dataset3.loc[(dataset3['phq9']>=15) & (dataset3['phq9']<19), 'phq_grp'] = 3
dataset3.loc[dataset3['phq9']>=20,'phq_grp'] = 4

dataset3.head(30)

#método alternativo, porém e necessário x,y
#dataset3['phq_grp'] = np.where((dataset3.phq9 > 5.1) & (dataset3.phq9 < 10), 1,0) #fonte https://pt.stackoverflow.com/questions/436720/python-np-where-com-duas-condi%C3%A7%C3%B5es
#dataset3['phq_grp'] = np.where((dataset3.phq9 >=10) & (dataset3.phq9 <15),2,0)
```
